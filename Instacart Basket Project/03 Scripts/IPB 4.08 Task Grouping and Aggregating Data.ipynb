# 03. Aggregating data with transform

```python
#deriving new variable for maximum amount of orders for each user
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merged.head(15)
```

# 04. Creating Loyalty Flag

```python
#creating flag for loyalty customers
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merged['loyalty_flag'].value_counts()
```

```python
#checking loyalty flag stats
ords_prods_merged[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

# 05. Find Spending Habits of Loyalty Levels

```python
#finding average prices for users to compare loyalty levels
ords_prods_merged['Average_spending']= ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merged.head()
```

```python
#checking stats for Average_spending column
ords_prods_merged[['user_id', 'loyalty_flag', 'Average_spending']].head(60)
```

# 06. Creating Spending Flag

```python
#Creating Spending Flag
ords_prods_merged.loc[ords_prods_merged['Average_spending'] < 10, 'Spending_Flag'] = 'Low Spender'
ords_prods_merged.loc[ords_prods_merged['Average_spending'] >= 10, 'Spending_Flag'] = 'High Spender' 
```

```python
ords_prods_merged.head()
```

```python
ords_prods_merged['Spending_Flag'].value_counts()
```

# 07. Creating Customer Frequency Flag

```python
#deriving variable for median days since last order
ords_prods_merged['median_days_between_orders'] = ords_prods_merged.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merged[['user_id', 'median_days_between_orders']].head(50)
```

```python
#Creating Order_Frequency Flag
ords_prods_merged.loc[ords_prods_merged['median_days_between_orders'] > 20, 'Order_Frequency'] = 'Non-Frequent customer'
ords_prods_merged.loc[(ords_prods_merged['median_days_between_orders'] <= 20) &(ords_prods_merged['median_days_between_orders'] > 10), 'Order_Frequency'] = 'Regular Customer'
ords_prods_merged.loc[ords_prods_merged['median_days_between_orders'] <= 10, 'Order_Frequency'] = 'Frequent customer'
```

```python
#Checking Frequency Flag Stats
ords_prods_merged[['user_id', 'Order_Frequency']].head(20)
```

```python
ords_prods_merged['Order_Frequency'].value_counts()
```

# 08. Comparing Subset Stats to Full DF Stats

```python
#creating subset
df = ords_prods_merged[:1000000]
```

```python
#checking department order stats of subset
df.groupby('department_id').agg({'order_number' : ['mean', 'min', 'max']})
```

```python
#checking department order stats of full df
ords_prods_merged.groupby('department_id').agg({'order_number' : ['mean', 'min', max]})
```

The min and max of the full set and the subset are the same because the minimum is always 1 for this data set and python caps the output for max at 99 as a default.  The main diffences are that there is a row for every department because in the full set every department has sales data and the means are different taking into account the full data set.  Some of the means are quite different eg.dept 17 which is ~15.7 int he subset vs ~ 11.3 in the full set but some are almost identical eg dept 7 which is 17.4 in the subset and 17.2 in the main set.  For mins and maxes the difference between the subset and full set are inconsiquential, but for averages taking the full set into account provides a much better picture of the full dataframe.
